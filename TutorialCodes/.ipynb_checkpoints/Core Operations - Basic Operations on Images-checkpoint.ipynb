### Paddings

```python
BLUE = [255,0,0]

img1 = cv2.imread('messi.jpg')

replicate = cv2.copyMakeBorder(img1,1000,100, 1000,1000,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img1,1000, 1000, 1000,1000,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img1,1000,1000,1000,1000,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img1,1000,1000,1000,1000,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,1000,1000,1000,1000,cv2.BORDER_CONSTANT,value=BLUE)

plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')

plt.show()
```

Writing images to a file (specify img type - jpg, png, etc)

```python
cv2.imwrite('messi_replicate.jpg', replicate)
cv2.imwrite('messi_reflect_101.jpg', reflect101)
cv2.imwrite('messi_constant.jpg', constant)
```
